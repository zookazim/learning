## File Manipulation

```python
import pandas as pd
```

## 1. Load Data from CSV

```python
full_file_path_input = 'c:/Users/he157349/source/repos/learning/python/pandas/admi.csv'
full_file_path_output = 'c:/Users/he157349/source/repos/learning/python/pandas/admi_modified.csv'
```

```python
# load data from csv file with header row and no index column
# If the file is not found, it will raise a FileNotFoundError   

df = pd.read_csv(full_file_path_input, header=0, index_col=None)
```

## 2. Show file info

```python
# Show header of the DataFrame
df.head()
```

```python
# Show header row in a list
header_row = df.columns.tolist()
print(header_row)
```

## 3. Remove Columns

```python
# Remove columns HitHNBeds, histHitHNBeds
df.drop(columns=['HitHNBeds', 'histHitHNBeds'], inplace=True)
```

## 4. Rename existing columns

```python
# Rename column histHITHNBeds to histHitHNBeds
df.rename(columns={'histHITHNBeds': 'histHitHNBeds'}, inplace=True)
```

```python
# Rename column histHITHNBeds to HitHNBeds
df.rename(columns={'HITHNBeds': 'HitHNBeds'}, inplace=True)
```

## 5. Check Results of file manipulation

```python
# Show header of the DataFrame
df.head()
```

```python
# Show header row in a list
header_row = df.columns.tolist()
print(header_row)
```

## 6. Save Results to File

```python
# Save df to file
df.to_csv(full_file_path_output, index=False, header=True)
```
