### 2.3 Create mapping

```python
mapping = {}
for fact in fact_fields:
    for submission in submission_fields:
        mapping[fact['name']] = submission['name']

print(mapping)
```

```python
mapping_list = []
for fact, submission in zip(fact_fields, submission_fields):
    mapping_list.append({"from": submission, "to": fact})

print(mapping_list)
```
